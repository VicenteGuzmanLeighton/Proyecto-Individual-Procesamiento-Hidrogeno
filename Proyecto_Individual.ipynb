**Introducción**

Actualmente el almacenamiento de energías renovables es necesario debido a que existe un desfase entre el consumo y la generación eléctrica. Una opción que resuelve la intermitencia de las energías renovables es el almacenamiento químico, donde la energía es almacenada a través de adsorción física, química o transformaciones químicas. Sus ventajas son la alta densidad energética y capacidad de almacenamiento por largo plazo.

El hidrógeno (H2) es un portador de energía y se puede almacenar químicamente en enlaces químicos dentro de compuestos mediante reacciones reversibles. Un ejemplo de esto es el compuesto químico Amoníaco (NH3), gas incoloro con sabor amargo que se licua a –33,3 ◦C y se congela hasta formar un cristal a −77,7 ◦C [1].

Por otro lado, la reacción de amoníaco con ácido nítrico produce una pastilla porosa, una sustancia altamente usada como fertilizante  conocida como nitrato de amonio (NH4NO3). Además tiene otros usos como explosivos en la minería, en refrigeradores y aires acondicionados, en la fabricación de ácidos, fibras, papeles, plásticos, polímeros y combustibles alternativos [2].

**2) Principio de operación de la unidad de proceso**

**Descripción del Principio de Operación del Proceso Haber-Bosch**

El proceso Haber-Bosch se basa en la conversión de nitrógeno (N2) e hidrógeno (H2) en amoníaco (NH3) mediante una reacción exotérmica que ocurre en condiciones controladas de alta presión y temperatura, y en presencia de un catalizador (generalmente a base de hierro) [5]. La reacción general es:

N2 (g) + 3H2 (g) ⇌ 2NH3 (g) + calor

**Funcionamiento Técnico del Proceso**

Para que el proceso Haber-Bosch sea eficiente, se requiere manejar varios parámetros técnicos críticos, lo que convierte esta unidad de proceso en un sistema complejo. A continuación se describen los principios operativos fundamentales y se incluye un esquema básico para facilitar la comprensión:

**1) Obtención y Compresión de los gases de reacción**:

El proceso comienza con la obtención de los gases de
reacción a partir de diversas fuentes.

En el caso del Hidrógeno, están:

- Reformado de Gas Natural
-  Electrólisis del Agua
- Reformado Autotérmico o Reformado de Biomasa
- Gasificación del Carbón

En el caso del Nitrógeno están:
- Separación Criogénica del Aire
- Separación por Membranas del Aire
- Absorción por Cambios de Presión (PSA)

De modo que el aire es el principal recurso para obtener nitrógeno, mientras que el gas natural actualmente se emplea para obtener hidrógeno [1]. Estos gases se purifican y se mezclan en las proporciones adecuadas para la reacción química.

Luego, estos gases se comprimen  a presiones elevadas de entre 150–300 atm para garantizar que la reacción ocurra bajo condiciones favorables. Una mezcla homogénea y comprimida adecuadamente aumenta la conversión de reactivos en NH3 [1].

**2) Calentamiento y Condiciones de Reacción**: La reacción de síntesis de amoníaco es exotérmica, pero requiere inicialmente una temperatura alta (400-500 °C) para superar la energía de activación y favorecer la velocidad de reacción en presencia del catalizador.

**3) Síntesis de amoníaco**: El proceso de síntesis de amoníaco se realiza en un reactor que contiene un catalizador de hierro, generalmente mezclado con promotores (como potasio o calcio). Este es fundamental para acelerar la reacción sin consumirse en el proceso. El catalizador permite que la reacción sea viable a una velocidad adecuada en condiciones industriales debido a que facilita la ruptura de los enlaces triples de la molécula de nitrógeno (N2) y reduce sustancialmente la energía de activación. Cabe destacar que el catalizador no se consume en la reacción y permanece disponible para repetir el ciclo de adsorción y reacción.

**4) Recirculación de Gases No Reaccionados**: Los gases no convertidos en amoníaco en el primer paso (principalmente nitrógeno e hidrógeno) se recirculan en el sistema para maximizar la eficiencia de uso de reactantes y reducir pérdidas.

**5) Enfriamiento y Separación**: Tras la reacción, la mezcla de gases pasa por un sistema de enfriamiento que permite condensar el amoníaco producido en forma líquida, mientras que los gases no reaccionados se separan y retornan al reactor en un sistema cerrado.

**3) Desarrollo del modelo matemático o termodinámico**


El modelo termodinámico del proceso Haber-Bosch en estado estacionario considera fundamentalmente un balance de masa y uno de energía junto con dos conceptos termodinámicos, la constante de equilibrio(K) y la velocidad de reacción.

El balance de masa global del proceso es:

**(F_N2,inicio) + (F_H2​,inicio) = (F_N2) + (F_H2) + (F_NH3)**

El balance de energía del proceso incluye todas las etapas del proceso como la compresión, calentamiento, reacción, recirculación y enfriamiento:

**(Q_in) + (W_in) = (Q_out) + (W_out) + ΔH reacción**

Donde:

- (Q_in): Corresponde al calentamiento de los gases

Q = ∑(F_i) * (Cp,i) * (T_out​−T_in)

- (W_in): Corresponde a la compresión de los gases

W = (n)* (R) * (T) * (ln(P_out/P_in))

- (Q_out): Corresponde al calor retirado para enfriamiento del amoníaco

Q​ = (F_NH3) * (​Cp_NH3) * (T_enf − T_out)

- (W_out): corresponde al trabajo adicional requerido para la recirculación de los gases.

- (ΔH): Corresponde al calor liberado por la reacción

La reacción que ocurre en el reactor es exotérmica, es decir, libera calor. El efecto de la temperatura sobre los equilibrios implica un cambio en la constante de equilibrio (K), que está determinado por el signo de ΔH de la reacción. La base teórica de esta dependencia está dada por la ecuación de Van't Hoff y Gibbs.

N2(g) + 3H2(g) ⇌ 2NH3(g)

**Ecuación de Gibss**: ΔGº = −R​⋅T​⋅ln(K)

**Ecuación de Van't Hoff**:  ln(K) =(−ΔH)/(R​⋅T)

**Reación de Equilibrio**: K  = (P_NH3)^2 / ((P_N2)^1)⋅((P_H2)^3)

 Al aumentar la temperatura disminuye la constante de equilibrio K ,ya que ΔH es negativo e igual a -92.4 (kJ/mol) [1].

 Por el principio de Le Chatelier, si un sistema en equilibrio se somete a un cambio en presión, temperatura o concentración, el equilibrio se desplazará en la dirección que tienda a contrarrestar ese cambio. Al aumentar la presión, como la reacción hacia los productos produce menos moles de gas (2 moles de NH3 frente a 4 moles de N2 y H2), el sistema favorece la formación de amoníaco, desplazándose hacia la derecha.

 Sin embargo, este fenómeno químico no ocurre de igual forma con la temperatura. A temperatura ambiente, el equilibrio favorece al amoníaco, pero la reacción no avanza a una velocidad apropiada debido a la alta energía de activación requerida. Bajar la temperatura no es útil porque la velocidad de reacción depende de la energía cinética de las moléculas, que aumenta con la temperatura, facilitando la colisión efectiva entre moléculas de N2 y H2. Además que el catalizador utilizado para ello requiere una temperatura de al menos 400 °C para ser eficiente [3].

A temperaturas entre 400 y 500 °C, se logra una velocidad de reacción considerable sin desplazar el equilibrio demasiado hacia los reactivos, lo que permite una producción razonable de NH3. Temperaturas superiores a 500 °C desplazarían el equilibrio hacia los reactivos, disminuyendo la cantidad de amoníaco formado y requiriendo un ciclo adicional de compresión y reacción, lo que implica costos energéticos adicionales


Esto se puede ver en el modelo interactivo a continuación.

```python
# Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Configuración de constantes
R = 8.314  # Constante de los gases ideales, J/(mol·K)
Delta_H_rxn = -92_220  # Calor de reacción en J/mol (exotérmico)
Delta_S_rxn = -198.0  # Entropía de reacción en J/(mol·K)

# 1. Ecuación de Van 't Hoff para la constante de equilibrio K
def constant_equilibrium(T):
    """
    Calcula la constante de equilibrio K para una temperatura T (en Kelvin).
    """
    return np.exp((-Delta_H_rxn / (R * T)) + (Delta_S_rxn / R))

# 2. Ecuación de velocidad de reacción (simplificada para ilustración)
def reaction_rate(P_N2, P_H2, P_NH3, K):
    """
    Calcula la velocidad de reacción en función de las presiones parciales y la constante de equilibrio.
    """
    k_forward = 1e3  # Constante de velocidad directa (asumida, para escala)
    return k_forward * (P_N2 * P_H2**3 - (P_NH3**2 / K))

# 3. Graficar constante de equilibrio K como función de la temperatura
temperatures_C = np.linspace(0, 700, 500)  # Temperaturas en °C
temperatures_K = temperatures_C + 273.15  # Convertir a Kelvin
K_values = [constant_equilibrium(T) for T in temperatures_K]  # Constantes de equilibrio

# Crear gráfico
plt.figure(figsize=(8, 6))
plt.plot(temperatures_C, K_values, label='Constante de equilibrio (K)', color='blue')
plt.xlim(300, 600)  # Rango de interés en °C
#plt.ylim(0, max(K_values))
plt.ylim(1e-6, 1e-2)
plt.xlabel('Temperatura (°C)', fontsize=12)
plt.ylabel('Constante de equilibrio (K)', fontsize=12)
plt.title('Dependencia de la constante de equilibrio con la temperatura', fontsize=14)
plt.legend()
plt.grid()
plt.show()

# 4. Balances de masa y energía
"""
¿Qué significa x?
- 'x' es el avance de reacción, una variable que describe la cantidad de sustancia (en moles) que ha reaccionado
  en un sistema químico. En el contexto del proceso Haber-Bosch:
    N2 + 3H2 ↔ 2NH3

- Relación estequiométrica:
  - Por cada mol de N2 que reacciona, se consumen 3 moles de H2 y se producen 2 moles de NH3.

¿De qué depende el valor de x?
- Temperatura (T): Influye en el equilibrio químico (constante K) y la velocidad de reacción.
- Presión (P): A mayor presión, se favorece la formación de NH3 según el principio de Le Châtelier.
- Relación molar inicial de N2 y H2: La proporción ideal es 1:3 (1 mol de N2 por 3 moles de H2).
- Eficiencia del catalizador: Afecta la velocidad de reacción y, por ende, la cantidad de conversión.
- Recirculación de gases: En la práctica, los gases no convertidos se recirculan para aumentar la conversión global.
"""
def mass_balance(F_N2_0, F_H2_0, x):
    """
    Calcula los flujos molares de salida del reactor.
    """
    F_N2 = F_N2_0 - x
    F_H2 = F_H2_0 - 3 * x
    F_NH3 = 2 * x
    return F_N2, F_H2, F_NH3

def energy_balance(T_in, T_out, P_in, P_out, F_N2_0, F_H2_0, x):
    """
    Calcula el balance energético global del proceso.
    """
    # Calor para calentar gases
    Cp_N2 = 29.12  # J/(mol·K)
    Cp_H2 = 28.84  # J/(mol·K)
    Cp_NH3 = 35.06  # J/(mol·K)
    Q_cal = F_N2_0 * Cp_N2 * (T_out - T_in) + F_H2_0 * Cp_H2 * (T_out - T_in)

    # Trabajo de compresión
    n = F_N2_0 + F_H2_0
    W_comp = n * R * T_in * np.log(P_out / P_in)

    # Calor de reacción
    Q_reac = -Delta_H_rxn * x

    # Energía para enfriar amoníaco
    F_NH3 = 2 * x
    T_cool = 298.15  # Temperatura ambiente en K
    Q_cool = F_NH3 * Cp_NH3 * (T_cool - T_out)

    # Balance global
    Q_in = Q_cal + W_comp
    Q_out = Q_reac + Q_cool
    return Q_in, Q_out

# Ejemplo interactivo (acá se pueden cambiar los valores)
F_N2_0 = 1.0  # mol/s de nitrógeno
F_H2_0 = 3.0  # mol/s de hidrógeno
x = 0.5  # Avance de reacción
T_in = 300 + 273.15  # Temperatura de entrada en K
T_out = 500 + 273.15  # Temperatura de salida en K
P_in = 1e5  # Presión de entrada en Pa
P_out = 2e6  # Presión de salida en Pa

# Calcular balances
Q_in, Q_out = energy_balance(T_in, T_out, P_in, P_out, F_N2_0, F_H2_0, x)
F_N2, F_H2, F_NH3 = mass_balance(F_N2_0, F_H2_0, x)

# Imprimir resultados
print("=== Resultados del modelo ===")
print(f"Flujo de N2 saliente: {F_N2:.2f} mol/s")
print(f"Flujo de H2 saliente: {F_H2:.2f} mol/s")
print(f"Flujo de NH3 producido: {F_NH3:.2f} mol/s")
print(f"Energía total de entrada: {Q_in:.2f} J/s")
print(f"Energía total de salida: {Q_out:.2f} J/s")

```

**4) Simulación del proceso bajo distintas condiciones y Evaluación del rendimiento de la Unidad de Proceso**

La simuación de la unidad de proceso se basa en maximizar el rendimiento de la reacción de producción de amoníaco mediante el ajuste o cambio en las condiciones de reacción. Entendiendo rendimiento como el procentaje de conversión o amoníaco producido.

Estas condiciones operacionales son fundamentalmente dos, la presión y temperatura, sin embargo, cabe destacar que existen otras que tambien afectan a la unidad de proceso y la reacción.

Cuando ocurre una reacción reversible en un sistema cerrado, el equilibrio dinámico se alcanza cuando:

a) la velocidad de la reacción directa es igual a la velocidad de la reacción inversa

b) la concentración de reactivos y productos permanece igual.

Sin embargo, como en todos los sistemas en equilibrio, la introducción de un cambio en el sistema (por ejemplo, cambiar la temperatura o la presión) moverá la posición del equilibrio en la dirección que se opone al cambio y, por lo tanto, afectará el rendimiento de la reacción.

**1. Efecto sobre el rendimiento si se reduce la temperatura**

La reacción directa es exotérmica. Por lo tanto, al disminuir la temperatura, se desplazará la posición del equilibrio hacia la derecha (para promover la reacción exotérmica generadora de calor) para oponerse al cambio y, por lo tanto, aumentar el rendimiento.

**2. Efecto sobre el rendimiento si se aumenta la presión.**

Se forman menos moles de gas en el lado derecho de la ecuación ( reaccionan 4 moles de gas y se forman 2 moles de gas ). Por lo tanto, al aumentar la presión, se moverá la posición del equilibrio hacia la derecha (para promover la formación de menos moles de gas) para oponerse al cambio y, por lo tanto, aumentar el rendimiento.

**3. Eliminación continua del producto amoníaco (mediante enfriamiento y licuefacción)**

La eliminación continua de amoníaco mediante licuefacción moverá la posición del equilibrio hacia la derecha (para reemplazar el amoníaco eliminado) para oponerse al cambio y, así, aumentar el rendimiento.

**4. Reciclaje continuo de gases de nitrógeno e hidrógeno que no han reaccionado**

El reciclaje continuo de N2  y H2  que no han reaccionado moverá la posición del equilibrio hacia la derecha (para consumir el exceso de reactivos) para oponerse al cambio y, así, aumentar el rendimiento.

El efecto sobre el rendimiento al modificar la presión y la temperatura se muestra gráficamente a continuación, destacando dos patrones.

a) A temperatura constante, el rendimiento de amoníaco aumenta con el aumento de la presión

b) A presión constante, el rendimiento de amoníaco disminuye con el aumento de la temperatura.

En el proceso Haber, todos los reactivos y productos son gaseosos. Por lo tanto, si se aumenta la presión de la mezcla de reacción , manteniendo la temperatura constante, aumentará la cantidad de moles por unidad de volumen de todos los reactivos y productos gaseosos y, por lo tanto, aumentará la velocidad de reacción a través de una mayor concentración.

El efecto de la temperatura en el rendimiento de la unidad de  proceso se puede ver en el código interactivo a continuación.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve, exp

# Constantes
R = 8.314  # Constante de los gases ideales (J/mol·K)
Temp0 = 273.15  # Conversión a Kelvin
pressures = [25, 50, 100, 200, 400]  # Presiones en bar (equivalente a atmósferas), (acá se pueden cambiar los valores)
x = symbols('x')  # Variable simbólica

# Rango de temperatura(ºC)
Temp_min = 0 + Temp0
Temp_max = 700 + Temp0
dTemp = 25
Temp_range = np.arange(Temp_min, Temp_max + dTemp, dTemp)

# Resultados
results = {P: [] for P in pressures}  # Diccionario para almacenar concentraciones

# Cálculo del equilibrio para cada presión
for P in pressures:
    for T in Temp_range:
        # Calcular Kp basado en datos hipotéticos de DeltaH y DeltaS
        DeltaH = -92.4e3  # Cambio de entalpía estándar (J/mol)
        DeltaS = -198.4  # Cambio de entropía estándar (J/mol·K)
        DeltaG = DeltaH - T * DeltaS
        Kp = np.exp(-DeltaG / (R * T))

        # Resolver ecuación de equilibrio [6]
        eq = Kp * P**2 * 27 * (1 - x)**4 - 16 * (x * (2 - x))**2
        sol = solve(eq, x)

        # Seleccionar solución física válida (entre 0 y 1)
        for s in sol:
            if s.is_real and 0 <= s <= 1:
                results[P].append((T - Temp0, float(s / (2 - s))))
                break  # Tomamos la primera solución válida

# Graficar resultados
plt.figure(figsize=(10, 6))
for P in pressures:
    temps, concs = zip(*results[P])  # Extraer temperaturas y concentraciones
    plt.plot(temps, concs, label=f'P = {P} bar')

# Etiquetas y configuración del gráfico
plt.xlabel('Temperatura [°C]')
plt.ylabel('Porcentaje de NH3 producido (%)')
plt.title('Conversión de gases a NH3 vs Temperatura (Isobaras)')
plt.legend(title='Presiones')
plt.grid(True)
plt.show()
```

Ahora, para visualizar el efecto de la presión se tiene el siguiente código interactivo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve, exp

# Constantes
R = 8.314  # Constante de los gases ideales (J/mol·K)
Temp0 = 273.15  # Conversión a Kelvin
x = symbols('x')  # Variable simbólica

# Rango de presión
P_min = 0
P_max = 400
dP = 10
P_range = np.arange(P_min, P_max + dP, dP)

# Temperaturas constantes (isotermas) en °C (acá se pueden cambiar los valores)
temperatures = [350, 400, 450, 500, 550]

# Resultados
results = {T: [] for T in temperatures}  # Diccionario para almacenar conversiones

# Cálculo del equilibrio para cada temperatura
for T_C in temperatures:
    T = T_C + Temp0  # Convertir a Kelvin
    for P in P_range:
        if P == 0:  # Evitar división por cero
            results[T_C].append(0)
            continue

        # Calcular Kp basado en datos hipotéticos de DeltaH y DeltaS
        DeltaH = -92.4e3  # Cambio de entalpía estándar (J/mol)
        DeltaS = -198.4  # Cambio de entropía estándar (J/mol·K)
        DeltaG = DeltaH - T * DeltaS
        Kp = np.exp(-DeltaG / (R * T))

        # Resolver ecuación de equilibrio [6]
        eq = Kp * P**2 * 27 * (1 - x)**4 - 16 * (x * (2 - x))**2
        sol = solve(eq, x)

        # Seleccionar solución física válida (entre 0 y 1)
        for s in sol:
            if s.is_real and 0 <= s <= 1:
                results[T_C].append(float(s / (2 - s)) * 100)  # Conversión a porcentaje
                break
        else:
            results[T_C].append(0)  # Si no hay solución válida, asignar 0

# Graficar resultados
plt.figure(figsize=(10, 6))
for T_C in temperatures:
    plt.plot(P_range, results[T_C], label=f'T = {T_C}°C')

# Etiquetas y configuración del gráfico
plt.xlabel('Presión [bar]')
plt.ylabel('Conversión de NH3 [%]')
plt.title('Conversión de gases a NH3 vs Presión (Isotermas)')
plt.legend(title='Temperaturas')
plt.grid(True)
plt.show()
```

Con estos graficos se puede ver:

- El primer gráfico muestra cómo la conversión a amoníaco disminuye con el aumento de la temperatura para presiones constantes, debido al equilibrio termodinámico desfavorable a altas temperaturas.

- El segundo muestra cómo la conversión de amoníaco aumenta con la presión para temperaturas constantes.

El rango óptimo de producción en la unidad de proceso es de 400ºC y 150-200 bar. Logrando la mayor conversión (15-20%) de forma que al recirclular los gases se genera de manera continua.


**5) Conclusiones y recomendaciones para la sociedad**

El proceso Haber-Bosch es una tecnología clave para la producción de amoníaco, un insumo esencial en la fabricación de fertilizantes que sustenta la agricultura moderna y la seguridad alimentaria global. A pesar de su relevancia histórica y actual, este proceso presenta desafíos importantes en términos de sostenibilidad, debido a su alta dependencia de combustibles fósiles y a las emisiones significativas de CO2 asociadas a la producción de hidrógeno como materia prima.

En el contexto chileno e internacional, la implementación de mejoras tecnológicas en el proceso Haber-Bosch, como la incorporación de hidrógeno verde producido a partir de fuentes renovables, representa una oportunidad estratégica para reducir la huella de carbono de la industria de fertilizantes e incorporar la alternativa de almacenamiento energético. Chile, con su abundante recurso solar y eólico, se posiciona como un actor relevante en la producción de hidrógeno verde y su integración en procesos industriales como el Haber-Bosch.

**Propuesta de Aplicaciones Concretas**

1) Producción de fertilizantes sostenibles: Integrar plantas de Haber-Bosch con unidades de electrólisis que generen hidrógeno a partir de energía renovable, maximizando la competitividad del amoníaco bajo en carbono.

2) Exportación de amoníaco verde: Aprovechar la capacidad de Chile para producir hidrógeno verde para exportar amoníaco como vector energético y fertilizante sostenible.

3) Simbiosis industrial: Fomentar parques industriales que combinen la producción de amoníaco con industrias que utilicen CO2 capturado como insumo, reduciendo emisiones netas.

4) Generación de energía descentralizada: Explorar el uso del amoníaco como vector energético en comunidades aisladas para almacenamiento y transporte de energía, reduciendo la dependencia de combustibles fósiles.

- Financiamiento
